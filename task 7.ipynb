```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Connect to SQLite Database
conn = sqlite3.connect("sales_data.db")

# Step 2: Write and execute an SQL query to get total quantity sold and revenue per product
query = """
SELECT 
    product, 
    SUM(quantity) AS total_qty, 
    SUM(quantity * price) AS revenue
FROM 
    sales
GROUP BY 
    product
"""

# Load the query result into a pandas DataFrame
df = pd.read_sql_query(query, conn)

# Step 3: Print the results
print("Sales Summary by Product:")
print(df)
```

```python
# Step 4: Plot a bar chart for revenue per product
df.plot(kind='bar', x='product', y='revenue', legend=False)
plt.title("Revenue by Product")
plt.xlabel("Product")
plt.ylabel("Revenue ($)")
plt.tight_layout()

# Optional: Save the chart as an image file
plt.savefig("sales_revenue_chart.png")

# Step 5: Show the chart
plt.show()
```

```python
# Close the database connection
conn.close()
```
